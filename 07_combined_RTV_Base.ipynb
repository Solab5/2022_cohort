```python
def rename_columns(df, column_mappings):
    """
    rename columns passing in a mapping, of new and old namings. new column comes first
    """
    for new_col, old_col in column_mappings:
        df[new_col] = df[old_col]
```

## Outlier Management and merging

```python
dataframes_list = [rukungiri_rtv_base, kyenjojo_rtv_base, kanungu_rtv_base, kagadi_standar_rtv_base]  

processed_dfs = [manage_outliers(df) for df in dataframes_list]
```

```python
# Combine all processed DataFrames
final_rtv_base_livestock = pd.concat(processed_dfs, ignore_index=True)
```
