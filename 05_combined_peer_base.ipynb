```python
def load_and_process_sheet(file_path, sheet_name, header, id_range):
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=header)
    df['ID'] = pd.Series(range(1, len(df) + 1))
    return df[(df['ID'] > id_range[0]) & (df['ID'] < id_range[1])]
```

```python
def process_all_sheets(file_path):
    sheet_configs = {
        'kyenjojo': ('kyenjojo_cntrl', 5, (2, 492)),
        'mitooma': ('mitooma_Cntrl.', 5, (9, 821)),
        'kaliro': ('kaliro control', 5, (10, 969)),
        'rubanda': ('rubanda cntrl', 5, (7, 661)),
        'kanungu': ('Kanungu Cntrl', 5, (5, 435))
    }

    dataframes = {}
    for name, (sheet, header, id_range) in sheet_configs.items():
        df = load_and_process_sheet(file_path, sheet, header, id_range)
        dataframes[name] = df     
        mean = df['HH Income + Consumption + Residues/Day'].mean()
        print(f"{name} mean: {mean}")

    return dataframes


```

```python
def process_all_sheets(file_path):
    sheet_configs = {
        'kyenjojo': ('kyenjojo_cntrl', 5, (2, 492)),
        'mitooma': ('mitooma_Cntrl.', 5, (9, 821)),
        'kaliro': ('kaliro control', 5, (10, 969)),
        'rubanda': ('rubanda cntrl', 5, (7, 661)),
        'kanungu': ('Kanungu Cntrl', 5, (5, 435))
    }
    
    dataframes = {}
    for name, (sheet, header, id_range) in sheet_configs.items():
        df = load_and_process_sheet(file_path, sheet, header, id_range)
        
        # Add the new column for Kanungu
        if name == 'kanungu':
            df["Livestock Asset Value"] = df["Livestock Asset Value (Ugx)"]        
        dataframes[name] = df     
        mean = df['HH Income + Consumption + Residues/Day'].mean()
        print(f"{name} mean: {mean}")
```

```python
def merge_dataframes(dataframes, columns_to_use):
    return pd.concat([df[columns_to_use] for df in dataframes.values()], ignore_index=True)
```
